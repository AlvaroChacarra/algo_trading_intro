### Solucion guiada

```python
class Order:
    def __init__(self, symbol, side, price, size):
        self.symbol = symbol
        self.side = side
        self.price = price
        self.size = size

    def notional(self):
        return self.price * self.size

    def describe(self):
        return f"{self.side} {self.size:.2f} {self.symbol} @ {self.price}"

btc_order = Order("BTCUSDT", "buy", 99980, 0.10)
eth_order = Order("ETHUSDT", "buy", 3520, 1.40)
```

## 7. Define `Trade`

**Practicas:** otra clase con logica propia.

Define `Trade` con los mismos atributos y un metodo `cash_flow()` que:

- reste dinero si `side == 'buy'`
- sume dinero si `side == 'sell'`


````python
# Escribe aqui
````

````python
if "Trade" not in globals():
    print("Te falta definir `Trade`.")
else:
    trade_probe = Trade("BTCUSDT", "buy", 100000, 0.10)
    print("cash_flow buy:", trade_probe.cash_flow())
````

### Solucion guiada

```python
class Trade:
    def __init__(self, symbol, side, price, size):
        self.symbol = symbol
        self.side = side
        self.price = price
        self.size = size

    def cash_flow(self):
        signed = -1 if self.side == "buy" else 1
        return signed * self.price * self.size
```


## 8. Crea `PositionTracker`

**Practicas:** estado interno.

Define `PositionTracker()` con:

- `cash = 0.0`
- `position = 0.0`
- `apply_trade(trade)`
- `equity(mark_price)`


````python
# Escribe aqui
````

````python
if "PositionTracker" not in globals():
    print("Te falta definir `PositionTracker`.")
else:
    tracker_probe = PositionTracker()
    print("cash:", tracker_probe.cash)
    print("position:", tracker_probe.position)
````

### Solucion guiada

```python
class PositionTracker:
    def __init__(self):
        self.cash = 0.0
        self.position = 0.0

    def apply_trade(self, trade):
        if trade.side == "buy":
            self.cash = self.cash - trade.price * trade.size
            self.position = self.position + trade.size
        else:
            self.cash = self.cash + trade.price * trade.size
            self.position = self.position - trade.size

    def equity(self, mark_price):
        return self.cash + self.position * mark_price
```
