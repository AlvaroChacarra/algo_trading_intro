````python
if "Trade" not in globals():
    print("Te falta definir `Trade`.")
else:
    trade_probe = Trade("BTCUSDT", "buy", 100000, 0.10)
    print("cash_flow buy:", trade_probe.cash_flow())
````

### Solucion guiada

```python
class Trade:
    def __init__(self, symbol, side, price, size):
        self.symbol = symbol
        self.side = side
        self.price = price
        self.size = size

    def cash_flow(self):
        signed = -1 if self.side == "buy" else 1
        return signed * self.price * self.size
```


## 8. Crea `PositionTracker` con estado privado

**Practicas:** estado interno, atributos privados.

En Lesson 1 (ejercicio 9) manejaste `cash` y `position` como variables sueltas.  
Ahora `PositionTracker` encapsula ese mismo estado — nadie puede tocarlo sin pasar por `apply_trade()`.

Define `PositionTracker()` con:

- `_cash = 0.0` (privado: guión bajo = convención Python de 'no tocar desde fuera')
- `_position = 0.0` (privado)
- `apply_trade(trade)` — modifica `_cash` y `_position`
- `equity(mark_price)` — calcula y devuelve el equity
- `__repr__()` — devuelve algo legible como `PositionTracker(cash=-1000.00, pos=0.0100)`

**Por qué guión bajo:** es la señal Python de 'este atributo es interno'.  
No lo protege técnicamente, pero comunica al equipo que no deben acceder directamente.

````python
# Escribe aqui
````

````python
if "PositionTracker" not in globals():
    print("Te falta definir `PositionTracker`.")
else:
    tracker_probe = PositionTracker()
    print("cash:", tracker_probe.cash)
    print("position:", tracker_probe.position)
````

### Solucion guiada

```python
class PositionTracker:
    def __init__(self):
        self._cash = 0.0       # privado: guión bajo
        self._position = 0.0

    def apply_trade(self, trade):
        if trade.side == "buy":
            self._cash -= trade.price * trade.size
            self._position += trade.size
        else:
            self._cash += trade.price * trade.size
            self._position -= trade.size

    def equity(self, mark_price):
        return self._cash + self._position * mark_price

    def __repr__(self):
        return f"PositionTracker(cash={self._cash:.2f}, pos={self._position:.4f})"
```

## 10b. Bonus — `MultiAssetTracker` (composición)

**Practicas:** composición (una clase que contiene otras clases).

Crea `MultiAssetTracker` que internamente use un `dict` de `{symbol: PositionTracker}`.  
Debe tener:

- `apply_trade(trade)` — delega en el tracker correcto según `trade.symbol`
- `total_equity(marks)` — recibe `{'BTCUSDT': 100090, 'ETHUSDT': 3540}` y devuelve la suma total
- `__repr__()` — lista los activos gestionados

**Por qué esto importa:** cuando tienes N activos, añadir uno nuevo es **una línea** en el dict.  
Compare con el script caótico del ejercicio 0: allí era un bloque entero duplicado.


````python
# Escribe aqui
````

````python
if 'MultiAssetTracker' not in globals():
    print('Te falta definir MultiAssetTracker.')
else:
    try:
        mat = MultiAssetTracker()
        mat.apply_trade(Trade('BTCUSDT', 'buy',  100000, 0.10))
        mat.apply_trade(Trade('ETHUSDT', 'buy',  3520,   1.40))
        marks = {'BTCUSDT': 100080, 'ETHUSDT': 3540}
        eq = mat.total_equity(marks)
        print('Bien: MultiAssetTracker funciona.')
        print('repr:', repr(mat))
        print('total_equity:', round(eq, 2))
        # BTC: cash=-10000 + 0.10*100080 = 8.0
        # ETH: cash=-4928 + 1.40*3540 = 28.0
        # total = 36.0
    except Exception as e:
        print(f'Error: {e}')
````

### Solucion guiada

```python
class MultiAssetTracker:
    def __init__(self):
        self._trackers = {}     # dict de {symbol: PositionTracker}

    def apply_trade(self, trade):
        if trade.symbol not in self._trackers:
            self._trackers[trade.symbol] = PositionTracker()
        self._trackers[trade.symbol].apply_trade(trade)

    def total_equity(self, marks):
        return sum(
            tracker.equity(marks[symbol])
            for symbol, tracker in self._trackers.items()
            if symbol in marks
        )

    def __repr__(self):
        symbols = list(self._trackers.keys())
        return f'MultiAssetTracker(activos={symbols})'
```

**Clave:** `MultiAssetTracker` no duplica lógica. Delega en `PositionTracker` y los organiza en un dict.  
Añadir SOLUSDT es automático: la primera vez que llega un trade de SOL, se crea el tracker.
