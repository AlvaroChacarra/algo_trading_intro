## 8. Cancela una orden

**Practicas:** construir una nueva lista filtrando elementos.

Después elimina la orden `2` y guarda el resultado en `active_book`.

````python
if "cancel_order_by_id" not in globals():
    print("Te falta definir `cancel_order_by_id`.")
elif not callable(cancel_order_by_id):
    print("`cancel_order_by_id` existe, pero no es una función.")
elif "active_book" not in globals():
    print("Te falta crear `active_book`.")
else:
    remaining_ids = [order['id'] for order in active_book]
    if remaining_ids != [1, 3, 4]:
        print("`active_book` debería quedarse con los ids [1, 3, 4].")
    else:
        print("Correcto")
````

### Solución guiada

```python
def cancel_order_by_id(order_book, order_id):
    new_book = []

    for order in order_book:
        if order['id'] != order_id:
            new_book.append(order)

    return new_book

active_book = cancel_order_by_id(order_book, 2)
print(active_book)
```

## 10. Ejecuta la estrategia toy

**Practicas:** combinar bucle, estado y reglas de trading muy simples.

**Pista:** recorre ambas listas usando el índice.

---

**Pregunta de cierre (no hay que codificarla):**

Imagina que quieres añadir ETH a esta estrategia — ejecutar señales en BTC **y** ETH a la vez.

Con el código actual necesitarías:
- Variables separadas: `eth_cash`, `eth_position`, `eth_equity`
- Duplicar la lógica del bucle para ETH
- Una función `total_equity` que sume BTC + ETH

¿Ves el patrón? Con tres activos serían tres bloques duplicados.  
En la siguiente clase aprenderás la herramienta que elimina esa duplicación: las **clases**.


````python
signals = ['buy', 'buy', 'flat', 'sell', 'buy', 'sell']
market_prices = [100000, 100040, 100030, 99990, 100080, 100120]

cash = 0
position = 0
trade_count = 0

# Escribe aquí
````

````python
missing = [name for name in ["trade_count", "cash", "position", "final_equity"] if name not in globals()]

if missing:
    print("Te faltan estas variables:", ", ".join(missing))
elif trade_count != 4:
    print("`trade_count` debería valer 4.")
elif cash != 30:
    print("`cash` debería valer 30.")
elif position != 0:
    print("`position` debería terminar en 0.")
elif final_equity != 30:
    print("`final_equity` debería valer 30.")
else:
    print("Correcto")
````

### Solución guiada

```python
for i in range(len(signals)):
    signal = signals[i]
    price = market_prices[i]

    if signal == 'buy' and position == 0:
        cash = cash - price
        position = 1
        trade_count = trade_count + 1
    elif signal == 'sell' and position == 1:
        cash = cash + price
        position = 0
        trade_count = trade_count + 1

final_equity = cash + position * market_prices[-1]

print('trade_count:', trade_count)
print('cash:', cash)
print('position:', position)
print('final_equity:', final_equity)
```
