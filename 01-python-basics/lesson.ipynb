## 2. Variables y tipos: guardar datos con nombre

Una variable permite reutilizar un valor, y Python distingue entre enteros, decimales y texto.

**Antes de ejecutar:** ¿qué tipo tendrá `symbol`? ¿Y `mid`?


```python
bid = 99950
ask = 100000
size = 0.25
symbol = "BTCUSDT"

spread = ask - bid
mid = (bid + ask) / 2
reference_notional = ask * size

print("symbol:", symbol)
print("spread:", spread)
print("mid:", mid)
print("reference_notional:", reference_notional)

print("type(symbol) ->", type(symbol))
print("type(bid)    ->", type(bid))
print("type(size)   ->", type(size))
print("type(mid)    ->", type(mid))
```

## 3. Listas y `for`: cuando ya no tienes un dato, sino muchos

Una lista guarda varios valores ordenados y un `for` permite aplicar la misma operación a todos.

**Antes de ejecutar:** ¿la media saldrá por encima o por debajo de `100000`?


```python
mid_prices = [99975, 99980, 99970, 99990, 100005]

print("all prices:", mid_prices)
print("first price:", mid_prices[0])
print("last price:", mid_prices[-1])

total = 0
for price in mid_prices:
    total = total + price

average_price = total / len(mid_prices)
print("average price:", average_price)
```

## 4. Diccionarios: una orden tiene piezas con nombre

Cuando un dato tiene campos con significado, un diccionario es más natural que una lista.

Aquí ya estamos modelando algo reconocible de trading.


```python
order = {
    "id": 1,
    "symbol": symbol,
    "side": "buy",
    "price": bid,
    "size": 0.10,
}

order["notional"] = order["price"] * order["size"]

print(order)
print("order side:", order["side"])
print("order price:", order["price"])
print("order notional:", order["notional"])
```

## 5. `if`: convertir una observación en una decisión

Un algoritmo no solo calcula: también clasifica y decide.

**Antes de ejecutar:** con este `spread`, ¿el mercado es `tight`, `normal` o `wide`?


```python
if spread <= 20:
    market_state = "tight"
elif spread <= 60:
    market_state = "normal"
else:
    market_state = "wide"

print("market_state:", market_state)
```

## 6. Funciones: empaquetar lógica para reutilizarla

Si una idea aparece más de una vez, merece un nombre y una función.

A partir de aquí el código empieza a parecerse más a piezas reutilizables que a comandos sueltos.


```python
def compute_mid(bid, ask):
    return (bid + ask) / 2


def compute_notional(price, size):
    return price * size


def classify_market(spread):
    if spread <= 20:
        return "tight"
    elif spread <= 60:
        return "normal"
    return "wide"


print("mid:", compute_mid(99950, 100000))
print("notional:", compute_notional(99980, 0.25))
print("state:", classify_market(50))
```
