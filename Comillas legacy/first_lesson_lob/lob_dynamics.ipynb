## 🔍 **1. Conceptos Fundamentales**

### 🤔 **¿Qué es un Limit Order Book?**

Un **Limit Order Book (LOB)** es una estructura de datos que mantiene todas las órdenes de compra y venta pendientes para un activo financiero específico, organizadas por precio y tiempo.

<div style="background-color: #f8f9fa; border-left: 5px solid #28a745; padding: 15px; margin: 10px 0; color: #2c3e50;">
<h4 style="color: #2c3e50;">🎯 <b>Definición Simple:</b></h4>
<p style="color: #2c3e50;">Imagina un <b>libro de órdenes</b> como una <b>subasta continua</b> donde:</p>
<ul style="color: #2c3e50;">
    <li>📘 <b>Lado COMPRA (Bid)</b>: Lista de compradores dispuestos a pagar diferentes precios</li>
    <li>📕 <b>Lado VENTA (Ask)</b>: Lista de vendedores dispuestos a vender a diferentes precios</li>
</ul>
</div>

### 🏗️ **Componentes Principales:**

| 🔧 **Componente** | 📝 **Descripción** | 🎨 **Visualización** |
|:---|:---|:---|
| **🟢 Bid Side** | Órdenes de compra (lado izquierdo) | Precios decrecientes ⬇️ |
| **🔴 Ask Side** | Órdenes de venta (lado derecho) | Precios crecientes ⬆️ |
| **💰 Best Bid** | Precio más alto de compra | Mayor precio en bid side |
| **💸 Best Ask** | Precio más bajo de venta | Menor precio en ask side |
| **📏 Spread** | Diferencia: Best Ask - Best Bid | "Costo de liquidez" |
| **📊 Mid Price** | (Best Bid + Best Ask) / 2 | Precio "justo" teórico |

### 🎭 **Analogía del Mercado:**

<div style="background: linear-gradient(90deg, #74b9ff, #0984e3); padding: 15px; border-radius: 8px; color: white;">
<h4>🏪 <b>Piensa en un Bazar Tradicional:</b></h4>
<ul>
    <li>🛒 <b>Compradores</b> gritan: "¡Compro manzanas a $2!"</li>
    <li>🍎 <b>Vendedores</b> gritan: "¡Vendo manzanas a $3!"</li>
    <li>💱 Una <b>transacción</b> ocurre cuando alguien acepta el precio del otro</li>
    <li>📋 El <b>LOB</b> es como un megáfono que organiza todos estos gritos</li>
</ul>
</div>

### ⚙️ **¿Cómo Funciona la Mecánica?**

```
🔄 FLUJO BÁSICO DEL LOB:

1️⃣ LLEGA ORDEN → 2️⃣ ¿PUEDE EJECUTARSE? → 3️⃣ SI: MATCH ✅ / NO: AL BOOK 📚

┌─────────────┐    ┌─────────────┐    ┌─────────────┐
│   ORDEN     │    │   MATCHING  │    │  RESULTADO  │
│   NUEVA     │───▶│   ENGINE    │───▶│   TRADE O   │
│             │    │             │    │   PENDING   │
└─────────────┘    └─────────────┘    └─────────────┘
```

<div style="background-color: #2c3e50; border: 1px solid #34495e; padding: 15px; border-radius: 5px; color: white;">
<h4 style="color: white;">📏 <b>¿Qué orden se ejecuta primero?</b></h4>
<ol style="color: white;">
    <li><b>🥇 PRECIO</b>: Mejor precio tiene prioridad</li>
    <li><b>🥈 TIEMPO</b>: Si mismo precio → Primero en llegar, primero en servirse (FIFO)</li>
</ol>

<p style="color: white;"><b>Ejemplo:</b> Si hay 3 órdenes de compra a $100, se ejecuta la que llegó primero ⏰</p>

## 🎯 ¿Cómo usar el simulador?

### 📝 **Limit Orders** (Órdenes Limitadas)
- Especifica **precio** y **cantidad**
- Se añade al LOB en el nivel de precio indicado
- **No se ejecuta inmediatamente** (a menos que cruce el spread)

### ⚡ **Market Orders** (Órdenes de Mercado)
- Solo especifica **cantidad** (el precio se deshabilita)
- **Se ejecuta inmediatamente** contra las mejores órdenes del lado opuesto
- Puede ejecutarse parcialmente si no hay suficiente liquidez

4. **Prueba cruzar el spread** con una limit order (ej: BUY a $101.5)

````python
# ⚡ ANÁLISIS DE VOLATILIDAD Y RIESGO

print("⚡ Calculando métricas de VOLATILIDAD y RIESGO...")
print("=" * 60)

# Calcular métricas de volatilidad
volatility_metrics = market_analyzer.calculate_volatility_metrics()

# Mostrar resultados detallados
if 'error' not in volatility_metrics:
    print("⚡ VOLATILITY & RISK ANALYSIS - RESULTADOS:")
    print("=" * 60)
    
    # Métricas básicas de volatilidad
    print(f"📈 Volatilidad Básica:")
    print(f"   • Realized Volatility: {volatility_metrics['realized_volatility']*100:.3f}%")
    print(f"   • Log Volatility: {volatility_metrics['log_volatility']*100:.3f}%")
    print(f"   • Annualized Volatility: {volatility_metrics['annualized_volatility']*100:.1f}%")
    print(f"   • Price Variability: ${volatility_metrics['price_variability']:.4f}")
    print(f"   • Spread Volatility: ${volatility_metrics['spread_volatility']:.4f}")
    
    # Métricas de riesgo ajustado
    print(f"\n📊 Risk-Adjusted Metrics:")
    print(f"   • Sharpe Ratio: {volatility_metrics['sharpe_ratio']:.3f}")
    print(f"   • Maximum Drawdown: {volatility_metrics['max_drawdown']*100:.2f}%")
    print(f"   • Volatility Stability: {volatility_metrics['volatility_stability']:.3f}")
    
    # Distribución de retornos
    print(f"\n📐 Distribución de Retornos:")
    print(f"   • Skewness: {volatility_metrics['skewness']:+.3f}")
    print(f"   • Kurtosis (excess): {volatility_metrics['kurtosis']:+.3f}")
    print(f"   • Volatility Clustering: {volatility_metrics['volatility_clustering']:+.3f}")
    
    # Interpretación profesional
    print(f"\n🎯 INTERPRETACIÓN DE RIESGO:")
    print("=" * 40)
    
    # Evaluar nivel de volatilidad
    annual_vol = volatility_metrics['annualized_volatility']
    if annual_vol < 0.15:  # 15%
        print("✅ VOLATILIDAD: Baja - Mercado estable")
    elif annual_vol < 0.30:  # 30%
        print("⚠️  VOLATILIDAD: Moderada - Riesgo normal")
    else:
        print("🚨 VOLATILIDAD: Alta - Mercado estresado")
    
    # Evaluar Sharpe ratio
    sharpe = volatility_metrics['sharpe_ratio']
    if sharpe > 1:
        print("✅ SHARPE: Excelente risk-adjusted performance")
    elif sharpe > 0:
        print("⚠️  SHARPE: Performance positiva pero moderada")
    else:
        print("🚨 SHARPE: Performance negativa")
    
    # Evaluar distribución
    skew = volatility_metrics['skewness']
    kurt = volatility_metrics['kurtosis']
    
    if abs(skew) < 0.5 and abs(kurt) < 2:
        print("✅ DISTRIBUCIÓN: Aproximadamente normal")
    elif abs(skew) < 1 and abs(kurt) < 5:
        print("⚠️  DISTRIBUCIÓN: Moderadamente no-normal")
    else:
        print("🚨 DISTRIBUCIÓN: Fuertemente no-normal - Riesgo de colas")
    
    # Evaluar drawdown
    max_dd = abs(volatility_metrics['max_drawdown'])
    if max_dd < 0.05:  # 5%
        print("✅ DRAWDOWN: Pérdidas controladas")
    elif max_dd < 0.15:  # 15%
        print("⚠️  DRAWDOWN: Pérdidas moderadas")
    else:
        print("🚨 DRAWDOWN: Pérdidas significativas")
        
else:
    print(f"❌ Error en cálculo de volatilidad: {volatility_metrics['error']}")

print("\n" + "=" * 60)
````

````python
# 📊 ANÁLISIS COMPLETO DE LA DISTRIBUCIÓN DE RETORNOS
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
from scipy import stats
import pandas as pd

# Generar datos de retornos simulados basados en las métricas calculadas
# Usaremos los parámetros de la crisis que ya tenemos
np.random.seed(42)  # Para reproducibilidad

# Simular retornos usando los parámetros del mercado
n_periods = 1000
base_vol = volatility_metrics['realized_volatility'] / np.sqrt(252)  # Vol diaria
mean_return = 0.0001  # Pequeño drift positivo

# Crear retornos con características no normales
returns = []
for i in range(n_periods):
    # Agregar clustering de volatilidad
    if i > 0 and abs(returns[-1]) > 2 * base_vol:
        vol_multiplier = 1.5  # Mayor volatilidad después de movimientos grandes
    else:
        vol_multiplier = 1.0
    
    # Generar return con skew y kurtosis
    random_shock = np.random.standard_t(df=5)  # t-distribution para fat tails
    daily_return = mean_return + base_vol * vol_multiplier * random_shock
    
    # Añadir sesgo ocasional (skew)
    if np.random.random() < 0.1:  # 10% de probabilidad
        daily_return += np.random.exponential(base_vol) * np.random.choice([-1, 1])
    
    returns.append(daily_return)

returns = np.array(returns)

# Calcular estadísticas descriptivas
mean_return = np.mean(returns)
std_return = np.std(returns)
skewness = stats.skew(returns)
kurt = stats.kurtosis(returns)
jarque_bera_stat, jarque_bera_p = stats.jarque_bera(returns)

# Test de normalidad Shapiro-Wilk (para muestras < 5000)
if len(returns) < 5000:
    shapiro_stat, shapiro_p = stats.shapiro(returns)
else:
    # Para muestras grandes usamos Kolmogorov-Smirnov
    ks_stat, ks_p = stats.kstest(returns, 'norm', args=(mean_return, std_return))

print("🔬 ANÁLISIS ESTADÍSTICO DE LA DISTRIBUCIÓN DE RETORNOS")
print("=" * 60)
print(f"📊 Número de observaciones: {len(returns):,}")
print(f"📈 Media: {mean_return:.6f}")
print(f"📊 Desviación estándar: {std_return:.6f}")
print(f"📐 Skewness: {skewness:.4f}")
print(f"📊 Kurtosis: {kurt:.4f}")
print(f"📊 Kurtosis exceso: {kurt - 3:.4f}")
print("\n🧪 TESTS DE NORMALIDAD:")
print("-" * 30)
print(f"🔬 Jarque-Bera test:")
print(f"   Estadístico: {jarque_bera_stat:.4f}")
print(f"   P-value: {jarque_bera_p:.6f}")
print(f"   ¿Normal?: {'NO' if jarque_bera_p < 0.05 else 'SÍ'}")

if len(returns) < 5000:
    print(f"🔬 Shapiro-Wilk test:")
    print(f"   Estadístico: {shapiro_stat:.4f}")
    print(f"   P-value: {shapiro_p:.6f}")
    print(f"   ¿Normal?: {'NO' if shapiro_p < 0.05 else 'SÍ'}")
else:
    print(f"🔬 Kolmogorov-Smirnov test:")
    print(f"   Estadístico: {ks_stat:.4f}")
    print(f"   P-value: {ks_p:.6f}")
    print(f"   ¿Normal?: {'NO' if ks_p < 0.05 else 'SÍ'}")

# Crear figura con subplots
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        '📊 Distribución de Retornos vs Normal',
        '📐 Q-Q Plot vs Distribución Normal', 
        '📈 Colas de la Distribución (Log Scale)',
        '🎯 Análisis de Outliers'
    ),
    specs=[[{"secondary_y": False}, {"secondary_y": False}],
           [{"secondary_y": False}, {"secondary_y": False}]]
)

# 1. Histograma con comparación normal
fig.add_trace(
    go.Histogram(
        x=returns,
        nbinsx=50,
        histnorm='probability density',
        name='Retornos Observados',
        marker_color='lightblue',
        opacity=0.7
    ),
    row=1, col=1
)

# Distribución normal teórica
x_normal = np.linspace(returns.min(), returns.max(), 100)
normal_dist = stats.norm.pdf(x_normal, mean_return, std_return)
fig.add_trace(
    go.Scatter(
        x=x_normal,
        y=normal_dist,
        mode='lines',
        name='Normal Teórica',
        line=dict(color='red', width=2)
    ),
    row=1, col=1
)

# 2. Q-Q Plot
sorted_returns = np.sort(returns)
theoretical_quantiles = stats.norm.ppf(np.linspace(0.01, 0.99, len(returns)), mean_return, std_return)

fig.add_trace(
    go.Scatter(
        x=theoretical_quantiles,
        y=sorted_returns,
        mode='markers',
        name='Q-Q Plot',
        marker=dict(color='blue', size=4, opacity=0.6)
    ),
    row=1, col=2
)

# Línea de referencia para Q-Q plot
min_val = min(theoretical_quantiles.min(), sorted_returns.min())
max_val = max(theoretical_quantiles.max(), sorted_returns.max())
fig.add_trace(
    go.Scatter(
        x=[min_val, max_val],
        y=[min_val, max_val],
        mode='lines',
        name='Línea 45°',
        line=dict(color='red', dash='dash')
    ),
    row=1, col=2
)

# 3. Análisis de colas (Log scale)
hist_values, bin_edges = np.histogram(returns, bins=50, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Filtrar valores positivos para log scale
positive_hist = hist_values[hist_values > 0]
positive_centers = bin_centers[hist_values > 0]

fig.add_trace(
    go.Scatter(
        x=positive_centers,
        y=positive_hist,
        mode='lines+markers',
        name='Densidad Observada',
        line=dict(color='blue'),
        marker=dict(size=4)
    ),
    row=2, col=1
)

normal_density = stats.norm.pdf(positive_centers, mean_return, std_return)
fig.add_trace(
    go.Scatter(
        x=positive_centers,
        y=normal_density,
        mode='lines',
        name='Densidad Normal',
        line=dict(color='red', dash='dash')
    ),
    row=2, col=1
)

# 4. Box plot para outliers
fig.add_trace(
    go.Box(
        y=returns,
        name='Retornos',
        boxpoints='outliers',
        marker_color='lightgreen'
    ),
    row=2, col=2
)

# Configurar ejes y layout
fig.update_xaxes(title_text="Retornos", row=1, col=1)
fig.update_yaxes(title_text="Densidad", row=1, col=1)

fig.update_xaxes(title_text="Cuantiles Teóricos", row=1, col=2)
fig.update_yaxes(title_text="Cuantiles Observados", row=1, col=2)

fig.update_xaxes(title_text="Retornos", row=2, col=1)
fig.update_yaxes(title_text="Densidad (Log Scale)", type="log", row=2, col=1)

fig.update_yaxes(title_text="Retornos", row=2, col=2)

fig.update_layout(
    height=800,
    title_text="🔬 Análisis Completo de la Distribución de Retornos",
    showlegend=True,
    template="plotly_white"
)

fig.show()

# Calcular percentiles para análisis de colas
percentiles = [0.1, 1, 5, 95, 99, 99.9]
observed_percentiles = np.percentile(returns, percentiles)
normal_percentiles = stats.norm.ppf(np.array(percentiles)/100, mean_return, std_return)

print(f"\n🎯 ANÁLISIS DE COLAS:")
print("-" * 30)
for i, p in enumerate(percentiles):
    obs = observed_percentiles[i]
    norm = normal_percentiles[i]
    ratio = abs(obs/norm) if norm != 0 else float('inf')
    print(f"Percentil {p:4.1f}%: Obs={obs:8.5f}, Normal={norm:8.5f}, Ratio={ratio:.2f}")

# Interpretación automática
print(f"\n🧠 INTERPRETACIÓN AUTOMÁTICA:")
print("=" * 50)

if abs(skewness) > 0.5:
    skew_direction = "derecha (positiva)" if skewness > 0 else "izquierda (negativa)"
    print(f"📐 SKEWNESS: SIGNIFICATIVO hacia la {skew_direction}")
    print(f"    Valor: {skewness:.3f} (>0.5 en valor absoluto)")
else:
    print(f"📐 SKEWNESS: MODERADO ({skewness:.3f})")

if kurt > 3:
    excess_kurt = kurt - 3
    if excess_kurt > 2:
        print(f"📊 KURTOSIS: COLAS MUY GRUESAS (fat tails)")
        print(f"    Exceso de kurtosis: {excess_kurt:.3f} (>>0)")
        print(f"    🚨 Riesgo de eventos extremos ALTO")
    else:
        print(f"📊 KURTOSIS: COLAS GRUESAS moderadas")
        print(f"    Exceso de kurtosis: {excess_kurt:.3f}")
elif kurt < 3:
    print(f"📊 KURTOSIS: COLAS DELGADAS ({kurt:.3f} < 3)")
else:
    print(f"📊 KURTOSIS: NORMAL ({kurt:.3f} ≈ 3)")

if jarque_bera_p < 0.05:
    print(f"🔬 NORMALIDAD: RECHAZADA (p-value: {jarque_bera_p:.6f})")
    print(f"    Los retornos NO siguen una distribución normal")
else:
    print(f"🔬 NORMALIDAD: NO RECHAZADA (p-value: {jarque_bera_p:.6f})")
    print(f"    Los retornos podrían seguir una distribución normal")

print(f"\n💡 IMPLICACIONES PARA TRADING:")
print("-" * 40)
if abs(skewness) > 0.5 or kurt > 5:
    print("⚠️  Usar modelos que capturen asimetría y colas gruesas")
    print("⚠️  VaR tradicional puede subestimar riesgo")
    print("⚠️  Considerar Expected Shortfall (CVaR)")
if jarque_bera_p < 0.05:
    print("⚠️  Los modelos basados en normalidad no son apropiados")
    print("⚠️  Usar distribuciones t de Student o skewed-t")
````
